```python
# Instalar bibliotecas necessárias
!pip install pandas matplotlib openpyxl gdown

import pandas as pd
import matplotlib.pyplot as plt
import gdown

# Função para baixar arquivo do Google Drive
def baixar_excel(id_arquivo, caminho_excel):
    """
    Baixa um arquivo Excel do Google Drive.
    :param id_arquivo: str - ID do arquivo no Google Drive.
    :param caminho_excel: str - Caminho para salvar o arquivo Excel.
    """
    url = f'https://drive.google.com/uc?id={id_arquivo}'
    gdown.download(url, caminho_excel, quiet=False)

# Função para carregar os dados do Excel
def carregar_dados(caminho_do_arquivo):
    """
    Carrega dados a partir de um arquivo Excel.
    :param caminho_do_arquivo: str - Caminho para o arquivo Excel.
    :return: DataFrame - Dados carregados.
    """
    dados = pd.read_excel(caminho_do_arquivo, header=1)  # Ajusta o cabeçalho para a linha correta
    return dados

# Função para realizar a análise dos dados
def analisar_dados(df):
    """
    Realiza a análise dos dados e exibe os resultados em gráficos.
    :param df: DataFrame - Dados processados.
    """
    # Exibir nomes das colunas
    print("Nomes das colunas no DataFrame:", df.columns.tolist())

    total_enem = df["Enem"].count()

    total_ac = df[df["Modalidade\nde Seleção"] == "AC"].shape[0]
    total_ac_apto = df[(df["Modalidade\nde Seleção"] == "AC") & (df["Resultado\nPreliminar"] == "Apto")].shape[0]
    total_ac_eliminado = df[(df["Modalidade\nde Seleção"] == "AC") & (df["Resultado\nPreliminar"] == "Eliminado")].shape[0]
    total_ac_inapto = df[(df["Modalidade\nde Seleção"] == "AC") & (df["Resultado\nPreliminar"] == "Inapto")].shape[0]

    total_li_ep = df[df["Modalidade\nde Seleção"] == "LI_EP"].shape[0]
    total_li_ep_apto = df[(df["Modalidade\nde Seleção"] == "LI_EP") & (df["Resultado\nPreliminar"] == "Apto")].shape[0]
    total_li_ep_eliminado = df[(df["Modalidade\nde Seleção"] == "LI_EP") & (df["Resultado\nPreliminar"] == "Eliminado")].shape[0]
    total_li_ep_inapto = df[(df["Modalidade\nde Seleção"] == "LI_EP") & (df["Resultado\nPreliminar"] == "Inapto")].shape[0]

    # Calcular porcentagens
    pct_ac_total_enem = (total_ac / total_enem) * 100
    pct_li_ep_total_enem = (total_li_ep / total_enem) * 100
    pct_total_modalidade_total_enem = ((total_ac + total_li_ep) / total_enem) * 100

    pct_ac_apto = (total_ac_apto / total_ac) * 100
    pct_ac_eliminado = (total_ac_eliminado / total_ac) * 100
    pct_ac_inapto = (total_ac_inapto / total_ac) * 100
    pct_ac_eliminado_inapto = ((total_ac_eliminado + total_ac_inapto) / total_ac) * 100

    pct_li_ep_apto = (total_li_ep_apto / total_li_ep) * 100
    pct_li_ep_eliminado = (total_li_ep_eliminado / total_li_ep) * 100
    pct_li_ep_inapto = (total_li_ep_inapto / total_li_ep) * 100
    pct_li_ep_eliminado_inapto = ((total_li_ep_eliminado + total_li_ep_inapto) / total_li_ep) * 100

    # Exibir resultados
    print("Total de inscritos em 'Enem':", total_enem)
    print("Total de AC em 'Modalidade de Seleção':", total_ac)
    print("Total de LI_EP em 'Modalidade de Seleção':", total_li_ep)
    print("Total 'Apto' referente ao AC em 'Resultado Preliminar':", total_ac_apto)
    print("Total 'Eliminado' referente ao AC em 'Resultado Preliminar':", total_ac_eliminado)
    print("Total 'Inapto' referente ao AC em 'Resultado Preliminar':", total_ac_inapto)
    print("Total 'Apto' referente ao LI_EP em 'Resultado Preliminar':", total_li_ep_apto)
    print("Total 'Eliminado' referente ao LI_EP em 'Resultado Preliminar':", total_li_ep_eliminado)
    print("Total 'Inapto' referente ao LI_EP em 'Resultado Preliminar':", total_li_ep_inapto)

    print("\nPorcentagens:")
    print(f"% Total de inscritos em 'Enem' por Total de AC em 'Modalidade de Seleção': {pct_ac_total_enem:.2f}%")
    print(f"% Total de inscritos em 'Enem' por Total de LI_EP em 'Modalidade de Seleção': {pct_li_ep_total_enem:.2f}%")
    print(f"% Total de inscritos em 'Enem' por Total de LI_EP em 'Modalidade de Seleção' + Total de AC em 'Modalidade de Seleção': {pct_total_modalidade_total_enem:.2f}%")

    print(f"% Total de AC em 'Modalidade de Seleção' por Total 'Apto' referente ao AC em 'Resultado Preliminar': {pct_ac_apto:.2f}%")
    print(f"% Total de AC em 'Modalidade de Seleção' por Total 'Eliminado' referente ao AC em 'Resultado Preliminar': {pct_ac_eliminado:.2f}%")
    print(f"% Total de AC em 'Modalidade de Seleção' por Total 'Inapto' referente ao AC em 'Resultado Preliminar': {pct_ac_inapto:.2f}%")
    print(f"% Total de AC em 'Modalidade de Seleção' por Total 'Eliminado' referente ao AC em 'Resultado Preliminar' + Total 'Inapto' referente ao AC em 'Resultado Preliminar': {pct_ac_eliminado_inapto:.2f}%")

    print(f"% Total de LI_EP em 'Modalidade de Seleção' por Total 'Apto' referente ao LI_EP em 'Resultado Preliminar': {pct_li_ep_apto:.2f}%")
    print(f"% Total de LI_EP em 'Modalidade de Seleção' por Total 'Eliminado' referente ao LI_EP em 'Resultado Preliminar': {pct_li_ep_eliminado:.2f}%")
    print(f"% Total de LI_EP em 'Modalidade de Seleção' por Total 'Inapto' referente ao LI_EP em 'Resultado Preliminar': {pct_li_ep_inapto:.2f}%")
    print(f"% Total de LI_EP em 'Modalidade de Seleção' por Total 'Eliminado' referente ao LI_EP em 'Resultado Preliminar' + Total 'Inapto' referente ao LI_EP em 'Resultado Preliminar': {pct_li_ep_eliminado_inapto:.2f}%")

    # Criar gráficos
    categorias = ['Apto', 'Eliminado', 'Inapto']
    valores_ac = [total_ac_apto, total_ac_eliminado, total_ac_inapto]
    valores_li_ep = [total_li_ep_apto, total_li_ep_eliminado, total_li_ep_inapto]

    fig, ax = plt.subplots()
    ax.bar(categorias, valores_ac, label='AC')
    ax.bar(categorias, valores_li_ep, bottom=valores_ac, label='LI_EP')
    ax.set_ylabel('Total')
    ax.set_title('Resultados por Modalidade de Seleção e Resultado Preliminar')
    ax.legend()
    plt.show()

# Função principal para rodar a análise do Excel
def main():
    """
    Função principal para rodar a análise dos dados do Excel.
    """
    # ID do arquivo Excel no Google Drive
    id_arquivo = '1BSArAFqD7wwUnCE1P1c1p-PbvSfNhP0F'
    caminho_excel = '/content/arquivo.xlsx'

    # Baixar arquivo Excel do Google Drive
    baixar_excel(id_arquivo, caminho_excel)

    # Carregar dados de exemplo
    df = carregar_dados(caminho_excel)

    # Analisar os dados
    analisar_dados(df)

# Executar a função principal
if __name__ == "__main__":
    main()
```
